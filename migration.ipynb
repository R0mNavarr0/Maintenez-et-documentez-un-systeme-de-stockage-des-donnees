```python
import pandas as pd
import kagglehub
```

```python
path = kagglehub.dataset_download("prasad22/healthcare-dataset")

df = pd.read_csv(path+'/healthcare_dataset.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Name'] = df['Name'].str.upper()
```

```python
df.head()
```

```python
df['Date of Admission'] =  pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] =  pd.to_datetime(df['Discharge Date'])
```

```python
df.info()
```

```python
df[['Prenom', 'Nom']] = df['Name'].str.split(' ',n=1, expand=True)

cols = ['Prenom', 'Nom'] + [col for col in df.columns if col not in ['Prenom', 'Nom']]
df = df[cols]
```
